```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot - MATLAB METHOD
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() - OBJECT ORIENTED INTERFACE METHOD 1
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() - OBJECT ORIENTED INTERFACE METHOD 2
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() - OBJECT ORIENTED INTERFACE METHOD 3
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot
ax = plt.axes()

# Legend = Boston
# Line width = 2 pt
# Green line
# Diamond Markers
ax.plot(x_axis, y_axis, label='Boston', linewidth=2, color='green', marker='D')
ax.legend()

# Grid lines
ax.grid()

# Title and axis labels
ax.set_title("Skill Drill 5.1.4")
ax.set_ylabel('Fare ($)')
ax.set_xlabel('Date')

# y-axis limit
ax.set_ylim(0, 45)
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(x_axis, y_axis)
```

```python
# Skill Drill 5.1.7
plt.plot(y_axis, x_axis, 'o', color='red', label = "Chicago")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("Skill Drill 5.1.7")
plt.legend()
plt.gca().invert_yaxis()
plt.xlim(0, 50)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Make all the dots larger
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# same thing without the for loop
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])

# "list comprehension" python specific feature
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Object Oriented Interface

# Change the size of the markers to 5 times each data point.
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*5)


# Add a black edge color to the circles.
# Make the linewidth of the circles 2 points.
# Change the color of the markers to sky blue.
# Add a legend for the city of Boston.
# Switch the axis so that the Fare($) data is on the x-axis.
# Make the color 20% transparent --  alpha = 0.2
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s = y_axis_larger, label="Boston", color = 'skyblue', alpha = 0.2, edgecolors='black', linewidth=2)
plt.legend()

# Add a title.
plt.title("Skill Drill 5.1.7")
plt.xlabel("Fare($)")
plt.ylabel("Date")

# Add a limit for the x-axis data.
plt.xlim(0, 50)

# Invert the y-axis so that January is at the top.
ax.invert_yaxis()

```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```
